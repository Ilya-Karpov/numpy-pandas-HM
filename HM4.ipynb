Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

-если источник traffic_source равен yandex или google, то в source_type ставится organic

-для источников paid и email из России - ставим ad

-для источников paid и email не из России - ставим other

-все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd
```

```python
log = pd.read_csv('доп_материалы_к_пандас_1_и_2/visit_log.csv', sep=';')
```

```python
log.loc[log['traffic_source'].isin(['yandex', 'google']),  'source_type'] = 'organic'
log.loc[(log['traffic_source'].isin(['paid', 'email'])) & (log['region'].str.contains('Russia')),  'source_type'] = 'ad'
log.loc[(log['traffic_source'].isin(['paid', 'email'])) & (~log['region'].str.contains('Russia')),  'source_type'] = 'other'
log.loc[~log['traffic_source'].isin(['yandex', 'google', 'paid', 'email']),  'source_type'] = log['traffic_source']
log.head(10)
```

Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
lifeline = pd.read_csv('доп_материалы_к_пандас_1_и_2/ml-latest-small/ratings.csv')
```

```python
morethen100 = lifeline.groupby('userId').count().reset_index()
more100 = morethen100[['userId' , 'movieId']]
```

```python
more100.loc[more100['movieId'] >= 100,  'lifeliner'] = '100+'
```

```python
only100 = more100.loc[~pd.isnull(more100.lifeliner), :]
```

```python
timestamp_max = lifeline.groupby('userId').max().reset_index()
timestamp_min = lifeline.groupby('userId').min().reset_index()
```

```python
short_max_timestamp = timestamp_max[['userId','timestamp']]
short_min_timestamp = timestamp_min[['userId','timestamp']]
```

```python
short_max_timestamp.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
short_min_timestamp.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
```

```python
avg_timestamp = only100.merge(short_max_timestamp, how = 'inner').merge(short_min_timestamp, how = 'inner')
```

```python
def average(row):
    return row['timestamp_max'] - row['timestamp_min']
```

```python
avg_timestamp['avg']=avg_timestamp.apply(average, axis=1)
avg_timestamp
```

```python
print (avg_timestamp['avg'].values.mean())
```
